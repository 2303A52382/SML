```python
#Questiion1
import pandas as pd
d=pd.read_excel('/content/dataset.xlsx')
print(d)
```

```python
y=d['Pass/Fail']
print(y)
```

```python
x=d.drop('Pass/Fail',axis=1)
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
m=lr.fit(x_train,y_train)
print(m.coef_)
print(m.intercept_)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
yp=m.predict(x_test)
print(accuracy_score(y_test,yp))
yp1=m.predict(x_train)
print(accuracy_score(y_train,yp1))
```

```python
print(confusion_matrix(y_test,yp))
```

```python
print(confusion_matrix(y_train,yp1))
```

```python
print(classification_report(y_test,yp))
```

```python
print(classification_report(y_train,yp1))
```

```python
data['Gender']=data['Gender'].map(gender)
data['Polyuria']=data['Polyuria'].map(polyuria)
data['Polydipsia']=data['Polydipsia'].map(polydipsia)
data['sudden weight loss']=data['sudden weight loss'].map(sudden_weight_loss)
data['weakness']=data['weakness'].map(weakness)
data['Polyphagia']=data['Polyphagia'].map(polyphagia)
data['Genital thrush']=data['Genital thrush'].map(genital_thrush)
data['visual blurring']=data['visual blurring'].map(visual_blurring)
data['Itching']=data['Itching'].map(itching)
data['Irritability']=data['Irritability'].map(irritability)
data['delayed healing']=data['delayed healing'].map(delayed_healing)
data['partial paresis']=data['partial paresis'].map(partial_paresis)
data['muscle stiffness']=data['muscle stiffness'].map(muscle_stiffness)
data['Alopecia']=data['Alopecia'].map(alopecia)
data['Obesity']=data['Obesity'].map(obesity)
data['class']=data['class'].map(class_col)
data
```

```python
x = data.drop('class', axis=1)
y = data['class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
m=lr.fit(x_train,y_train)
print(m.coef_)
print(m.intercept_)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
yp=m.predict(x_test)
print(accuracy_score(y_test,yp))
yp1=m.predict(x_train)
print(accuracy_score(y_train,yp1))
```
