```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Reading data

```python
d=pd.read_csv('/content/magic gamma telescope.csv')
d
```

Checking for null values

```python
d.isnull()
```

```python
mapping={"g":1,"h":2}
d['class']=d['class'].map(mapping)
print(d)
```

Target variable

```python
y=d['class']
print(y)
```

Features

```python
x=d.drop(['class'],axis=1)
print(x)
```

normalization

```python
dd1=(x-x.min())/(x.max()-x.min())
dd1
```

string to integer

train_test_split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,
test_size=0.2, random_state=42)
x_train
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
y_train
```

SVM classification

```python
from sklearn.svm import SVC
lr=SVC()
m=lr.fit(x_train,y_train)
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
yp=m.predict(x_test)
print(accuracy_score(y_test,yp))

```

```python
print(classification_report(y_test,yp))
```

```python
print(confusion_matrix(y_test,yp))
```

perceptron

```python
from sklearn.linear_model import Perceptron
lr1=Perceptron()
m=lr1.fit(x_train,y_train)

yp1=m.predict(x_test)
print(accuracy_score(y_test,yp1))
```

```python
print(confusion_matrix(y_test,yp1))
```

```python
print(classification_report(y_test,yp1))
```

logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr2=LogisticRegression(max_iter=1000)
m=lr2.fit(x_train,y_train)

yp2=m.predict(x_test)
print(accuracy_score(y_test,yp2))
```

```python
print(confusion_matrix(y_test,yp2))
```

```python
print(classification_report(y_test,yp2))
```
